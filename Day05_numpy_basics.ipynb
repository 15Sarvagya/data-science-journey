REORGANIZING ARRAYS

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
after = before.reshape(8,1)
print(after)
```

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
after = before.reshape((2,2,2))
print(after)
```

```python
#vertical stacking
v1=np.array([1,2,3])
v2=np.array([4,5,6])
np.vstack([v1,v2])
```

```python
#Horizontal stacking
h1=np.ones((2,4))
h2=np.zeros((2,2))
np.hstack([h1,h2])
```
