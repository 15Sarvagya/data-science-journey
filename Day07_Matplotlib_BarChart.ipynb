```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
labels = ['A','B','C']
y = [1,4,2]
bars=plt.bar(labels,y,color="y")
bars[0].set_hatch("/.")
bars[1].set_hatch("O/")
bars[2].set_hatch("-*")
plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ['A','B','C']
y = [1,4,2]
bars=plt.bar(labels,y,color="brown")
patterns=['/','*','o']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch("/.")
# bars[1].set_hatch("O/")
# bars[2].set_hatch("-*")
plt.figure(figsize=(6,4))

plt.show()
```
