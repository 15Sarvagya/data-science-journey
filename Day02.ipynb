```python
L=[1,2,4,-5,7,9,3,2]
for j in range(len(L)):
    m = L[j]
    idx = j
    c = j 
    for i in range(j,len(L)):
        if L[i]<m:
            m=L[i]
            idx=c
        c+=1
    temp = L[j]
    L[j] = m
    L[idx] = temp
print(L)

    
    
```

```python
def mysum(*args):
    s=0
    for i in range(len(args)):
        s +=args[i]
        i +=1
        
    return s
```
