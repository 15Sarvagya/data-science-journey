```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [1,2,3]
y= [2,4,6]
plt.plot(x,y)
plt.title("My First Graph",fontdict={'fontname':'Lucida Sans Unicode','fontsize':20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")


plt.show()
```

```python
x = [0,1,2,3]
y= [0,2,4,6]
plt.plot(x,y,color='green', label ='2x')
plt.title("My First Graph",fontsize=20,fontweight=100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='#ababab', label ='2x')
plt.show()
```

```python
x = [0,1,2,3]
y= [0,2,4,6]
plt.plot(x,y,color='green', label ='2x',linewidth =5,marker ='*',markersize=19,markeredgecolor="red",linestyle="--")
plt.title("My First Graph",fontsize=20,fontweight=100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
#ShortHand Notation(To make Graph Quick)
x = [0,1,2,3]
y= [0,2,4,6]
plt.plot(x,y,'r^--',label="2x")
plt.title("My First Graph",fontsize=20,fontweight=100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
# Adding Another Line in graph
x = [0,1,2,3]
y= [0,2,4,6]
plt.plot(x,y,'r^--',label="2x")
x2=np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'y*:',label="x^2")
plt.title("My First Graph",fontsize=20,fontweight=100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5])
# plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
#Future Projection

x = [0,1,2,3]
y= [0,2,4,6]
#Resize Graph
plt.figure(figsize=(5,3),dpi=100)
plt.plot(x,y,'r^--',label="2x")
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'y',label="x^2")
plt.plot(x2[5:],x2[5:]**2,'b--',)
plt.title("My First Graph",fontsize=20,fontweight=100)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5])
# plt.yticks([0,2,4,6,8,10])
plt.legend()
#To Save Graph
plt.savefig("My First Grapgh.png")
plt.show()
```
