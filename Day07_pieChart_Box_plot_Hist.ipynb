```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fifa = pd.read_csv(r"C:\Users\sarva\Downloads\fifa_data.csv")
fifa
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color='#593f13')
plt.xticks(bins)
plt.xlabel('Skills of players')
plt.ylabel('number of players')
plt.title('Skills of fifa players in 2018',fontdict={'fontsize':20})
plt.show()
```

```python
##PIE CHART

counts= fifa['Preferred Foot'].value_counts()
lables = ['Left','Right']
explodes=[0,0.03]

plt.pie([counts['Left'],counts['Right']],labels=lables,colors=['#f5b0f1','#3fc5d1'],autopct='%.2f %%',explode=explodes)
plt.title('Foot Prefernce of FIFA Player')
plt.show()
```

```python
counts= fifa['Preferred Foot'].value_counts()
lables = ['Left','Right']
explodes=[0,0.03]
plt.style.use('ggplot')

plt.pie([counts['Left'],counts['Right']],labels=lables,autopct='%.2f %%',explode=explodes)
plt.title('Foot Prefernce of FIFA Player')
plt.show()
```

```python
##BOXPLOT
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
england = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
tick_labels = ['FC','Madrid','England']

boxes=plt.boxplot([barcelona,Real_madrid,england],labels=tick_labels,patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#cc7ec8',linewidth = 2)
    box.set(facecolor='#e0e0e0')
            
            

plt.show()
```
